```python
#1.Tokenization
import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt_tab')
```

```python
from nltk.tokenize import word_tokenize

text ="Natural language processing is intresting"

tokens=word_tokenize(text)

print(tokens)
```

```python
#Sentence Tokenization
from nltk.tokenize import sent_tokenize

text ="Phyton is easy. It is widely used in data science."
sentences = sent_tokenize(text)
print(sentences)
```

```python
#Stop Word Removal
#Stop words such as is,the, and, in are removed..

from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
words = word_tokenize(text)
filtered_words = [w for  w in words if w.lower() not in stop_words]
print(filtered_words)
```

```python
#Stemming
#Stemming reduces words to their root form

from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
words = ["playing","studies","running","gone","watched","coding"]

for w in words:
  stemmed_word = stemmer.stem(w)
  print(stemmed_word)
```

```python
#Lemmatization
#Lemmatization converts words to their dictionary form.

import nltk
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

print(lemmatizer.lemmatize("writing"))
print(lemmatizer.lemmatize("studies"))
print(lemmatizer.lemmatize("best"))
print(lemmatizer.lemmatize("running"))
print(lemmatizer.lemmatize("coding"))
```

```python
#Part-of-Speech (POS) Tagging
#POS tagging identifies grammatical roles of words
import nltk
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
from nltk import pos_tag

words = word_tokenize("Phyton is a powerful programming language")
pos_tags = pos_tag(words)
print(pos_tags)
```

```python
#Word Frequency Count
from nltk.probability import FreqDist

words = word_tokenize(text)
fd = FreqDist(words)
print(fd.most_common())
```
